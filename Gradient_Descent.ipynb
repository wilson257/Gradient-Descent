```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_targets=1,n_informative=1,noise=20,random_state=13)
```

```python
plt.scatter(X,y);
```

```python
class GradientDescent:
    
    def __init__(self,learning_rate,epochs):
        self.b=100
        self.m=-120
        self.lr=learning_rate
        self.epochs=epochs
    
    def fit(self,Xtrain,ytrain):
        for i in range(self.epochs):
            slope_b= - 2 * np.sum(ytrain - self.m * Xtrain.ravel() - self.b)
            self.b= self.b - (self.lr * slope_b)
            
            slope_m= - 2 * np.sum((ytrain - self.m * Xtrain.ravel() - self.b) * Xtrain.ravel())
            self.m= self.m - (self.lr * slope_m)
            
            X_ip=np.linspace(-2,2,80)
            ypreds= self.m* X_ip + self.b
        plt.plot(X_ip,ypreds,c='red')
        plt.scatter(Xtrain,ytrain)
            
        print("Slope:",self.m," ","Intercept:",self.b)
        
    def predict(self,Xtest):
        y_pred= (self.m * Xtest) + self.b
        return y_pred
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
GD=GradientDescent(0.01,50)
```

```python
GD.fit(Xtrain,ytrain)
```

```python
GD.predict(Xtest)
```
